```
import pandas as pd
df_train = pd.read_csv('https://raw.githubusercontent.com/plenoi/CMU_DataScience/master/train.txt',sep='|',header=None)
```

```
#!pip install pythainlp
import pythainlp
train_clean = []
for i in range(0,len(df_train)):
    train_clean.append(pythainlp.tokenize.word_tokenize(df_train.iloc[i][0]))
```

```
train_clean[0]
```

```
train_clean_join = []
for i in range(0,len(train_clean)):
    train_clean_join.append(' '.join(train_clean[i]))
train_clean_join[0]
```

```
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(tokenizer=lambda x:x.split(),ngram_range = (1,2), stop_words =["ฉัน","ผม"],
                     min_df=10,max_df=0.9,norm=None,
                     use_idf=True,smooth_idf=True)
cv.fit(train_clean_join)
X = cv.transform(train_clean_join).toarray()
y = df_train[1]
X
```

```
cv.vocabulary_
```

```
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
len(cv.vocabulary_)
```

```
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import cross_val_score
clf = MultinomialNB()
scores = cross_val_score(clf,X_train,y_train, cv=10)
scores.mean()
```

```
clf = MultinomialNB()
clf.fit(X_train,y_train)
```

```
y_predict = clf.predict(X_test)
print("Test accuracy : "+str(sum(y_test == y_predict)/len(y_test)*100))
```
